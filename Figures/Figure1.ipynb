```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/decision_points.csv")

top_items = df["label_item"].value_counts().head(20)  #reduce to 15

plt.figure(figsize=(7, 5))  #smaller, doc-friendly
top_items.sort_values().plot(kind="barh")

plt.title("Top 20 Most Frequent Items", fontsize=10)
plt.xlabel("Count", fontsize=9)
plt.ylabel("Item", fontsize=9)

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

plt.tight_layout()

plt.savefig("item_distribution.png", dpi=300)  #high quality for docs
plt.show()
```
